```python
import numpy as np
import pandas as pd
```

```python
train_df = pd.read_csv('pm25_train.csv')
test_df = pd.read_csv('pm25_test.csv')
```

```python
def cleaningdata(data):
    data.drop(['date'],axis=1,inplace=True)
    return data
```

```python
x_train = cleaningdata(train_df)
x_test = cleaningdata(test_df)
```

```python
y_train = x_train['pm2.5']
```

```python
x_train.drop(['pm2.5'],axis=1,inplace=True)
```

```python

# corr plot## corr # corr p 
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
corr = train_df.corr()
f, ax = plt.subplots(figsize=(20, 20))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5)
plt.show()
```

```python
from sklearn.linear_model import Ridge

reg = Ridge(normalize=True)
reg.fit(x_train, y_train)
preds = reg.predict(x_test)
```

```python
kk = []
for i in preds:
    kk.append(round(i))
sub = pd.DataFrame({'pm2.5':kk}, columns=['pm2.5'])
sub.to_csv('Ridge_output.csv',index=None)
```
